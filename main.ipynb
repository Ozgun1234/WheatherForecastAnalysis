```python
# All the Null Values in the Data

df.isna().any()
```

```python
# Rename the column name Weather of the dataframe to Weather Condition

df.rename(columns={"Weather": "Weather Condition"}, inplace=True)

df.head()
```

```python
# Mean of Visibility

df.Visibility_km.mean()
```

```python
# Standard Deviation of Pressure

df.Press_kPa.std()
```

```python
# Column name formatting

df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ", "_")
```

```python
# All instances when Snow was recorded

df[df["weather_condition"].apply(lambda x : "Snow" in x)]
```

```python
# All instances when Wind Speed is above 24 & Visibility is 25

df[(df["wind_speed_km/h"] > 24) & (df["visibility_km"] == 25)]
```

```python
# Convert Date time column into proper data type.

df["date/time"] = pd.to_datetime(df["date/time"])
```
